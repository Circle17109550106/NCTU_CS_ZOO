# HW 5: 小鴨車作業系統安裝
2020.10.30

1. 下載 hypriot os (414 MB)
    * [hypriotos-rpi-v1.12.3.img.zip](https://github.com/hypriot/image-builder-rpi/releases/download/v1.12.3/hypriotos-rpi-v1.12.3.img.zip)
    * 將下載後的檔案 hypriotos-rpi-v1.12.3.img.zip 進行解壓縮，得到 hypriotos-rpi-v1.12.3.img
2. 下載燒錄工具，並安裝 (< 1min.)
    * Windows: [balenaEtcher](https://github.com/balena-io/etcher/releases/download/v1.5.109/balenaEtcher-Setup-1.5.109.exe)
    * Mac: [balenaEtcher](https://github.com/balena-io/etcher/releases/download/v1.5.109/balenaEtcher-1.5.109.dmg)
3. 燒錄 OS (3 min.)
    * 將記憶卡插入讀卡機中，連接電腦
    * 打開 balenaEtcher 燒錄程式
    * `Flash from file` 請選擇步驟 1 **解壓縮後的檔案** hypriotos-rpi-v1.12.3.img
    * `Select target` 請選擇對應的裝置
    > Note
    > 請確認選擇到 sd card ，**不要** 選到系統碟，您的筆電有可能變磚
    > **不要** 選到系統碟 ！！
    > **不要** 選到系統碟 ！！
    > **不要** 選到系統碟 ！！
    > 人家 `Etcher` 就告訴你隱藏，不要選麻
    > ![](https://i.imgur.com/BHxvFeY.png)

    * ![](https://i.imgur.com/EA3O33w.png)
    * 執行
        * Mac
            * 因為需要讀寫 block device 的權限，所以需要輸入密碼
    * 有任何疑慮，請聯繫助教
4. 設定 user-data.txt 及 config.txt (< 5 min.)
    * 燒錄完成後，拔出重新插入**電腦**
        * Windows
            * 打開檔案總管
            * 燒錄成功會看到一個 HypriotOS (X:)
                * ![](https://i.imgur.com/eFUVdd3.png)
        * Mac
            * 打開 Finder
            * 燒錄成功會看到一個 HypriotOS 在左邊
                * ![](https://i.imgur.com/SXKN9YB.png)
    * 打開後，找到 `user-data`，以 記事本 (Windows) 或 文字編輯器 (Mac) 開啟
        * Mac: 避免使用 **錯誤** 的 `Terminal.app` 開啟
    * 刪除 `user-data` 中的所有內容
    * 將 [our-user-data](https://github.com/GrassLab/NCTU_CS_ZOO/blob/Lab5/others/user-data.yaml) 中的內容複製後，貼進 `user-data`
    * 更改以下幾項內容
        1. 找到 `hostname: duckie0`，將 `duckie0` 改為 `duckie-學號`
            * Example: `hostname: duckie-0756049`
        2. 找到 `name: duckie`，將 `duckie` 更改為你想要的帳號名稱
            * 避免使用數字開頭以及中文的使用者名稱
            * Example: `name: handsome`
        3. 找到 `plain_text_passwd: "duckie"`，將 `duckie` 更改你想要的密碼，並牢牢記住
            * Example: `plain_text_passwd: "handsome"`
        4. 找到 `ssid="home_wifi_network"`，將 `home_wifi_network` 更改為你宿舍的無線網路名稱；找到 `psk="home_wifi_password"`，將 `home_wifi_password` 更改為你宿舍無線網路的密碼
            * Example: `ssid="my-homewifi"` `psk="12345678"`
            > Note
            > 如果沒有宿舍無線網路，請跳過第四步
            > NCTU 等學校提供的無線網路，因透過 WPA-EAP 驗證，因學校沒有提供正確設定檔給 pi，故無法使用
        5. 找到 `ssid="cellphone_wifi_network"`，將 `cellphone_wifi_network` 更改為你手機的無線網路名稱；找到 `psk="cellphone_wifi_password"`，將 `cellphone_wifi_password` 更改為你宿舍無線網路的密碼
            * Example: `ssid="my-phonewifi"` `psk="12345678"`
            > Note
            > 如果沒有手機無線網路，請跳過第四步

        > Note
        > 如果同時沒有宿舍無線網路及手機無線網路，請暫停安裝，並連繫助教
    * 存檔後關閉 user-data
    * 回到檔案管理系統，找到資料夾中的 `config.txt`（與 user-data 在同一個資料夾），以適當的文字編輯器開啟
    * 刪除 `config.txt` 中的所有內容
    * 將 [our-config](https://github.com/GrassLab/NCTU_CS_ZOO/blob/Lab5/others/boot_config.txt) 中的內容複製後，貼進 `config.txt`
    * 存檔後關閉 `config.txt`
    * 將記憶卡從電腦中退出
    * 有任何疑慮，請聯繫助教
5. 開機 (~ 5 min.)
    * 將記憶卡從讀卡機中拔出
    * 將小鴨車接上電源，**並點兩下行動電源的電源鍵關閉電源**
    * 將記憶卡插入小鴨車上的 RPi 3b+
    * 開啟行動電源
    * **第一次開機需要一段時間，等待約五分鐘**
6. 安裝 ssh 連線工具 (10 min.)
    * Windows
        * 下載 [MobaXterm](https://download.mobatek.net/2032020060430358/MobaXterm_Installer_v20.3.zip)
        * 解壓縮 MobaXterm_Installer_v20.3.zip 後，雙擊 MobaXterm_installer_20.3 以安裝 MobaXterm
        * 開啟 MobaXterm
        * 新增 Session
            * 選擇左上角的 Session
            * 點選 SSH
                * ![](https://i.imgur.com/TsQ1zkq.png)
                * 在 `Basic SSH settings` 中，`Remote host` 填入 `duckie-學號.local`
                    * Example: `duckie-0756049.local`
                * 在 `Basic SSH settings` 中，勾選 `Specify username`，並填入前面設定 user-data 時，你設定的 `name`
                    * Example: `handsome`
                * ![](https://i.imgur.com/4efkDaP.png)
                * 點選 `OK`
        * 完成後左方會出現前步驟所新增的 sessin，第一次新增完畢後會自動開啟，往後只要對其**雙擊**便可使用
            * ![](https://i.imgur.com/uMXX5pH.png)
        * 連上 session 後，左方選單中會自動多出一個 sftp 的連線，列出遠端的目錄結構，並且包含了 MobaTextEditor 文件編輯器，同學如果覺得 vim 編輯器不好使用，也可以透過 MobaTextEditor 做文件的編輯，文件在編輯存檔後，會自動上傳到遠端。
            * ![](https://i.imgur.com/pURnR2W.png)
            * 下方的 Follow terminal folder 可以勾起來，當右方 terminal 更換目錄時，左方的 sftp 目錄會跟著連動。
    * Mac
        * 下載 [Royal TSX](https://royaltsx-v4.royalapps.com/updates/royaltsx_4.3.4.1000.dmg)
        * 雙擊 `royaltsx_4.3.4.1000.dmg`，安裝 Royal TSX
        * 開啟 Royal TSX
            * 不使用 Licence，選擇繼續
            * 透過左上角 `Royal TSX` :arrow_right: `Plugins...` 打開
            ![](https://i.imgur.com/VsVGamC.jpg)
            * 安裝 `Terminal` Plugin
            * 安裝 `File Transfer` Plugin
            * 將 Royal TSX 關閉
        * 安裝 X Server
            * 下載 [XQuartz-2.7.11.dmg](https://dl.bintray.com/xquartz/downloads/XQuartz-2.7.11.dmg)
            * 雙擊後安裝，安裝後將其打開，啟動 X server，並成為一個視窗程式，但目前沒有任何 window 在裡面
            * 使用終端機打入，`launchctl load /Library/LaunchAgents/org.macosforge.xquartz.startx.plist` 與 `launchctl start org.macosforge.xquartz.startx`
                * 使用 `launchctl list | grep startx` 確認已經執行該 Agent
        * 重新開啟 Royal TSX
        * 新增 Terminal
            * 點 `File` :arrow_right: `New Document`
            ![](https://i.imgur.com/MJ11sfK.png)

            * 點選左下角的 `+` 號
            ![](https://i.imgur.com/cPaEBgB.png)
            * 選擇 Terminal
                * 在 `Computer Name` 填入 `duckie-學號.local`
                    * Example: `duckie-0756049.local`
                * 點到左邊的 `Credentials`，選擇 `Specify username and password`，在 `Username` 與 `Password` 填入前面設定的 user-data 時，你設定的 `name` 與 `plain_text_passwd`
                    * Example: `handsome`
                * 點到 `Advanced` 切換到 `SSH` 頁面，啟動 `X11 forwarding`
                ![](https://i.imgur.com/qQatRP0.jpg)
                * 點選 `Apply & Close`
        * 完成後左方會出現前步驟鎖好增的 Terminal，往後只要對其雙擊便可使用
        ![](https://i.imgur.com/RVDuyCw.png)
        * 連上 Terminal 後，可以點上方的 `Actions` :arrow_right: `Connect with Options` :arrow_right: `Connect using SFTP`，成功後，會列出在遠端的目錄結構。同學如果覺得 vim 編輯器不好使用，也可以點開檔案做文件的編輯，文件在編輯存檔後，會自動上傳到遠端
        ![](https://i.imgur.com/m7ksVdK.png)
        
    * Linux (Arch、Ubunut 等)
        * 請聯繫助教
7. 環境設定 (6 min.)
    * 在透過 ssh 連線工具連進小鴨車以後，安裝相關套件
       ```bash=
       cd
       git clone -b Lab5 https://github.com/GrassLab/NCTU_CS_ZOO.git Lab5
       cd Lab5
       sudo apt update
       sudo apt install -y $(cat others/apt-packages.txt)
       pip3 install -r duckiebot_cs_zoo/requirements.txt
       ```
8. 執行 lab5 測試程式
    * ```bash=
      cd ~/Lab5/duckiebot_cs_zoo
      python simple_joystick.py
      ```
    * You can use `w` `s` `a` `d` `space` to control the duckiebot.
